```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('preprocessed_data_immigration.txt')
df.head()
df.info()
```

```python
# En fazla geçen 20 kelimeyi bulma
top_words = df.iloc[:, 0].str.split(expand=True).stack().value_counts().head(20)
top_words

plt.figure(figsize=(10, 6))
bars = top_words.plot(kind='bar', color='skyblue')

for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 30, round(yval, 2), ha='center', va='bottom')

plt.title('En Fazla Geçen 20 Kelime')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 'cleaned_sentences' sütunundaki NaN değerleri boş bir dize ile değiştirin
df['cleaned_sentences'] = df['cleaned_sentences'].fillna('')

# 'cleaned_sentences' sütunundaki string tipinde olmayan değerleri temizleyin
df = df[df['cleaned_sentences'].apply(lambda x: isinstance(x, str))]

# Tüm cümleleri bir liste içinde toplayın
sentences = df['cleaned_sentences'].apply(lambda x: x.split()).tolist()

# Word2Vec modelini oluşturun
model = Word2Vec(sentences, min_count=1)

# Artık model üzerinde işlemlerinizi gerçekleştirebilirsiniz
similar_words = {}
for word in top_words.index:
    similar_words[word] = [similar_word for similar_word, _ in model.wv.most_similar(word, topn=5)]

print(similar_words)
```

```python
# 5 farklı cümle seçme
selected_sentences = df.sample(5)['cleaned_sentences'].tolist()
selected_sentences
#%%
# Her cümlenin en benzer 3 cümlesini bulma
similar_sentences = {}
for sentence in selected_sentences:
    # selected_sentences listesinden bulunan cümleyi çıkar
    df_without_selected = df[df['cleaned_sentences'] != sentence]

    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(df_without_selected['cleaned_sentences'])
    cosine_similarities = cosine_similarity(tfidf_vectorizer.transform([sentence]), tfidf_matrix).flatten()
```

```python
    # En benzer 3 cümleyi ve sıra numaralarını bul
    indices = cosine_similarities.argsort()[:-4:-1]
    similar_sentences[sentence] = [(idx, df_without_selected.loc[idx, 'cleaned_sentences']) for idx in indices]

similar_sentences
```
